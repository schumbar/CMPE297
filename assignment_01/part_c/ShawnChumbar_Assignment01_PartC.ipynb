### Image Generation

```python
import torch
from diffusers import StableDiffusionPipeline
from IPython.display import display, Image

def image_generation_demo():
    print("Image Generation Demo")

    # Initialize the Stable Diffusion pipeline
    model_id = "runwayml/stable-diffusion-v1-5"
    pipe = StableDiffusionPipeline.from_pretrained(model_id, torch_dtype=torch.float16)
    pipe = pipe.to("cuda")

    # Generate an image
    prompt = "A serene landscape with mountains and a lake at sunset"
    image = pipe(prompt).images[0]

    # Save the image
    image.save("generated_image.png")
    print("Image saved as 'generated_image.png'")

    # Display the image
    display(Image("generated_image.png"))

# Run the demo
image_generation_demo()
```

### Code Generation

```python
from transformers import GPT2LMHeadModel, GPT2Tokenizer

def code_generation_demo():
    print("Code Generation Demo")

    # Load pre-trained model and tokenizer
    model_name = "gpt2"
    model = GPT2LMHeadModel.from_pretrained(model_name)
    tokenizer = GPT2Tokenizer.from_pretrained(model_name)

    def generate_code(prompt, max_length=100):
        input_ids = tokenizer.encode(prompt, return_tensors="pt")
        output = model.generate(input_ids, max_length=max_length, num_return_sequences=1, no_repeat_ngram_size=2)
        generated_text = tokenizer.decode(output[0], skip_special_tokens=True)
        return generated_text

    prompts = [
        "def factorial(n):",
        "class BankAccount:",
        "def is_palindrome(s):"
    ]

    for prompt in prompts:
        print(f"\nPrompt: {prompt}")
        generated_code = generate_code(prompt)
        print("Generated code:")
        print(generated_code)

# Run the demo
code_generation_demo()
```

### Video Analysis

```python
import cv2
import torch
import torchvision.transforms as transforms
from torchvision.models import resnet50
from torchvision.models.resnet import ResNet50_Weights
from urllib.request import urlretrieve

def video_analysis_demo():
    print("Simplified Video Analysis Demo")
    video_filename = "/content/drive/MyDrive/Fall24/CMPE297_SpecTopics/Assignment_01/Part C/BigBuckBunny_320x180.mp4"

    # Load the pre-trained model
    model = resnet50(weights=ResNet50_Weights.IMAGENET1K_V2)
    model.eval()

    # Prepare the image transformer
    transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])

    # Open the video file
    cap = cv2.VideoCapture(video_filename)

    # Process every 30th frame
    frame_count = 0
    while cap.isOpened():
        ret, frame = cap.read()
        if not ret:
            break

        frame_count += 1
        if frame_count % 30 == 0:
            # Convert BGR to RGB
            frame_rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

            # Prepare the frame
            input_tensor = transform(frame_rgb)
            input_batch = input_tensor.unsqueeze(0)

            # Perform inference
            with torch.no_grad():
                output = model(input_batch)

            # Get the predicted class
            _, predicted_idx = torch.max(output, 1)
            predicted_label = ResNet50_Weights.IMAGENET1K_V2.meta["categories"][predicted_idx.item()]

            print(f"Frame {frame_count}: Predicted class - {predicted_label}")

    cap.release()
```

```python
video_analysis_demo()
```

### Multimodal Fusion

```python
import torch
import torchvision.transforms as transforms
from transformers import BertTokenizer, BertModel
from PIL import Image
import librosa
import numpy as np
import urllib.request
import torchvision
```

```python
audio_url = "https://upload.wikimedia.org/wikipedia/commons/f/f5/%27ariri.ogg"
text = "A beautiful sunset over the ocean"
image_url = "https://upload.wikimedia.org/wikipedia/commons/thumb/5/58/Sunset_2007-1.jpg/640px-Sunset_2007-1.jpg"
```

```python
def download_file(url, filename):
    urllib.request.urlretrieve(url, filename)
    print(f"Downloaded {filename}")

def multimodal_fusion_demo():
    print("\n10. Multimodal Fusion")
    # Download sample files
    image_file = "sunset.jpg"
    audio_file = "waves.ogg"
    download_file(image_url, image_file)
    download_file(audio_url, audio_file)

    # Process text
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
    model = BertModel.from_pretrained('bert-base-uncased')
    inputs = tokenizer(text, return_tensors="pt")
    outputs = model(**inputs)
    text_embedding = outputs.last_hidden_state.mean(dim=1)  # Shape: (1, 768)

    # Process image
    image_transform = transforms.Compose([
        transforms.Resize((224, 224)),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),
    ])
    image = Image.open(image_file)
    image_tensor = image_transform(image).unsqueeze(0)
    image_model = torchvision.models.resnet18(pretrained=True)
    image_embedding = image_model(image_tensor)  # Shape: (1, 1000)

    # Process audio
    audio, sr = librosa.load(audio_file, duration=10)
    mfccs = librosa.feature.mfcc(y=audio, sr=sr, n_mfcc=13)
    audio_embedding = torch.tensor(mfccs.mean(axis=1)).unsqueeze(0)  # Shape: (1, 13)

    # Ensure all embeddings have the same dimensions (2D tensors)
    text_embedding = text_embedding  # Already (1, 768)
    image_embedding = image_embedding  # Already (1, 1000)
    audio_embedding = audio_embedding  # Already (1, 13)

    # Print shapes for debugging
    print(f"Text embedding shape: {text_embedding.shape}")
    print(f"Image embedding shape: {image_embedding.shape}")
    print(f"Audio embedding shape: {audio_embedding.shape}")

    # Simple fusion: concatenation
    fused_embedding = torch.cat((text_embedding, image_embedding, audio_embedding), dim=1)
    print(f"Fused embedding shape: {fused_embedding.shape}")

    # Simple classification based on fused embedding
    classifier = torch.nn.Linear(fused_embedding.shape[1], 5)  # 5 arbitrary classes
    output = classifier(fused_embedding)
    predicted_class = torch.argmax(output, dim=1).item()
    print(f"\nPredicted class based on multimodal fusion: {predicted_class}")
```

```python
# Run the demo
multimodal_fusion_demo()
```
